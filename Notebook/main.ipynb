### Cambio de fecha rental

```python
import re
```

```python
def date(col):
    
    i = re.findall(r'\d+(?=[ -])', col)
    
    i = '-'.join(i)
    
    return i
```

```python
rental['rental_date'] = rental['rental_date'].apply(lambda x: date(x))
```

```python
rental['rental_date'] = rental.rental_date.astype(str)
```

```python
rental['return_date'] = rental['return_date'].apply(lambda x: date(x))
```

```python
rental['return_date'] = rental.return_date.astype(str)
```

```python
rental.rental_date.astype(str)
```

```python
rental.return_date.apply(str)
```

```python
rental['return_date'] = pd.to_datetime(rental['return_date'], format= '%Y-%m-%d')
```
